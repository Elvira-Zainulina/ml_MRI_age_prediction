```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import sklearn
from sklearn.decomposition import PCA
from sklearn.ensemble import ExtraTreesClassifier, RandomForestClassifier
from sklearn.manifold import TSNE, Isomap
from sklearn.metrics import accuracy_score, balanced_accuracy_score, f1_score
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold, cross_validate
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.svm import SVC, LinearSVC
from xgboost import XGBClassifier

from imblearn.over_sampling import SMOTE
```

```python
# change to path to your csv
path_to_unrestricted_csv = 'unrestricted_brazhnikovm_3_16_2020_12_57_48.csv'

data = pd.read_csv('unrestricted_hcp_freesurfer.csv', index_col='Subject')
data = data.drop(columns='Release')
target = pd.read_csv(path_to_unrestricted_csv, index_col='Subject')
target = target['Age']
data = data.join(target).dropna()
target = data['Age']
X = data.drop(columns='Age')

gender_labeling = LabelEncoder()
X['Gender'] = gender_labeling.fit_transform(X['Gender'])

target_labeling = LabelEncoder()
y = target_labeling.fit_transform(target)
y = y > 1
y = pd.Series(y, index=X.index)
```

# Try different algorithms

```python
def find_scores(model, X, y, cv):
    result = cross_validate(model, X, y, cv=cv, scoring=['f1_micro'])
    fsc = result['test_f1_micro']
    print('f1_micro: mean: {:.3f}, std: {:.3f}'.format(
    np.mean(fsc), np.std(fsc)))
```

```python
model = XGBClassifier(random_state=32)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=123)
find_scores(model, X, y, cv)
```

```python
model = ExtraTreesClassifier(random_state=32)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=123)
find_scores(model, X, y, cv)
```

```python
model = RandomForestClassifier(random_state=32)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=123)
find_scores(model, X, y, cv)
```

**xgboost has the highest mean score and the lowest variance so it will be used further**

# Use train-val-test split

```python
train_names = np.load('train_names.npy')
test_names = np.load('test_names.npy')
val_names = np.load('val_names.npy')
```

```python
X_train, X_test, y_train, y_test = X.loc[train_names], X.loc[test_names], y.loc[train_names], y.loc[test_names]
X_val, y_val =  X.loc[val_names], y.loc[val_names]
```

```python
model = XGBClassifier(random_state=32)
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
f1_score(y_test, pred, average='micro')
```

```python
best_score = 0
params = (3, 0.1, 100)
for max_depth in [1, 3, 5, 10]:
    for learning_rate in [0.1, 0.05, 0.01]:
        for n_estimators in [100, 200, 500]:
            model = XGBClassifier(max_depth=max_depth, learning_rate=learning_rate, n_estimators=n_estimators,
                                  random_state=32)
            model.fit(X_train, y_train)
            pred = model.predict(X_val)
            score = f1_score(y_val, pred, average='micro')
            if score > best_score:
                best_score = score
                params = (max_depth, learning_rate, n_estimators)
print(best_score, params)
```

```python
model = XGBClassifier(max_depth=params[0], learning_rate=params[1], n_estimators=params[2], random_state=32)
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
f1_score(y_test, pred, average='micro')
```

# Try SMOTE

```python
class SMOTE_XGBC(sklearn.base.BaseEstimator):
    
    def __init__(self, params):
        super().__init__()
        self.smote = SMOTE(random_state=32)
        self.model = XGBClassifier(max_depth=params[0], learning_rate=params[1], n_estimators=params[2], random_state=32)
        
    def fit(self, X, y):
        X_resampled, y_resampled = self.smote.fit_resample(X, y)
        self.model.fit(X_resampled, y_resampled)
    
    def predict(self, X):
        return self.model.predict(X)
```

```python
model = SMOTE_XGBC(params)
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
f1_score(y_test, pred, average='micro')
```

```python
xx = pred[:, 0]
yy = pred[:, 1]
for color, label in zip([False, True], ['<31', '>= 31']):
    plt.scatter(xx[y == color], yy[y == color], label=label)
plt.legend()
plt.title('Division by age group')
```
